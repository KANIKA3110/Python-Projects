```python
import tkinter as tk
from tkinter.font import BOLD
from tkinter import messagebox
import os
from time import sleep
from tkinter.constants import RIGHT, Y
import sqlite3
con=sqlite3.connect("empdb.db")
con.execute('''CREATE TABLE IF NOT EXISTS empd (ID INT PRIMARY KEY NOT NULL, NAME TEXT NOT NULL, POST TEXT NOT NULL, SALARY INT NOT NULL); ''')
```

```python
def check(id):

    a= con.execute("SELECT ID FROM empd")
    for row in a:
        if row[0]==id:
            return True
    return False



def main_screen() :
    main_window = tk.Tk()

    main_window.geometry("500x300")

    main_window.title("Employee Database")
    head = tk.Label(main_window, text="\nAdmin Portal\n", font=("Arial", 25)).pack()

    def to_proceed() :
        main_window.destroy()
        proceed()

    def exit() :
        main_window.destroy()
        


    proceed_button = tk.Button(text="Proceed", bg="green", fg="white", height=2, width=10, command=to_proceed)
    proceed_button.place(x=120, y=150)

    empty = tk.Label(text=" ").pack()

    exit_button = tk.Button(text="Exit", bg="red", fg="white", height=2, width=10, command=exit)
    exit_button.place(x=300, y=150)

    main_window.mainloop()

    
    
def proceed():
    global window2
    window2 = tk.Tk()

    window2.geometry("600x600")

    window2.title("Employee Database-Admin Portal")
    head = tk.Label(window2, text="\nSelect Operation\n", font=("Arial", 25)).pack()

    def to_proceed() :
        window2.destroy()
        proceed()

    def exit1() :
        window2.destroy()
        
    def back1() :
        window2.destroy()
        main_screen()

    add_button = tk.Button(text="Add Employee", bg="green", fg="white", height=2, width=15, command=Add)
    add_button.place(x=150, y=150)

    remove_button = tk.Button(text="Remove Employee", bg="green", fg="white", height=2, width=15, command=Remove)
    remove_button.place(x=350, y=150)

    promote_button = tk.Button(text="Promote Employee", bg="green", fg="white", height=2, width=15, command=Promote)
    promote_button.place(x=150, y=250)

    display_button = tk.Button(text="Display Employees", bg="green", fg="white", height=2, width=15, command=Display)
    display_button.place(x=350, y=250)

    
    
    back_button = tk.Button(text="Back", bg="blue", fg="white", height=1, width=10, command=back1)
    back_button.place(x=180,y=400)
    
    exit_button = tk.Button(text="Exit", bg="red", fg="white", height=1, width=10, command=exit1)
    exit_button.place(x=350, y=400)
    
    
    window2.mainloop()
    
    
    
def add(id, name, post, salary):
    
        data=(id, name, post, salary)
        sql = 'insert into empd (ID, NAME, POST, SALARY) values(?,?,?,?)'
           
        c=con.cursor()
        c.execute(sql,data)
        con.commit()
        print("Data added successfully:")
        proceed()

        
def Add() :
    window2.destroy()
    global awindow
    awindow = tk.Tk()
    awindow.geometry("400x350")
    awindow.title("Employee Database-Admin Portal")

    head = tk.Label(awindow, text="\nAdd New Employee Details\n", font=("Arial", 20)).pack()
    
    def to_add() :
        
        try:
            id=idtk.get()
            name=nametk.get()
            post=posttk.get()
            salary=salarytk.get()
        
            id=int(id)
            salary=int(salary)
            
        except:
            print("ID and Salary must be integer only!!")
            tk.messagebox.showwarning("Warning", "ID and Salary must be in integer format!!")
            Add()
            
        if check(id)==True:
                print("Employee with same ID already exists!!")
                tk.messagebox.showwarning("Warning", "Employee with same ID already exists!!")
                Add()
    
        else:
            tk.messagebox.showinfo("Information", "Employee added successfully!!")
            awindow.destroy()
            add(id,name,post,salary)
            
                
    def back3() :
        awindow.destroy()
        proceed()
        
    l_1 = tk.Label(awindow, text=" Unique ID ").place(x=70, y=100)
    
    idtk = tk.Entry(awindow, fg='blue', bg='white', width=30)
    idtk.place(x=140, y=100)

    l_2 = tk.Label(awindow, text="Name ").place(x=70, y=150)
   
    nametk = tk.Entry(awindow, fg='blue', bg='white', width=30)
    nametk.place(x=140, y=150)
    
    l_3 = tk.Label(awindow, text="Post ").place(x=70, y=200)
   
    posttk = tk.Entry(awindow, fg='blue', bg='white', width=30)
    posttk.place(x=140, y=200)
    
    l_4 = tk.Label(awindow, text="Salary ").place(x=70, y=250)
    
    salarytk = tk.Entry(awindow, fg='blue', bg='white', width=30)
    salarytk.place(x=140, y=250)

    Submit = tk.Button(awindow, text="Submit", bg="green", fg="white", command=to_add).place(x=190, y=300)
    
    back3_button = tk.Button(text="Back", bg="blue", fg="white", height=1, width=10, command=back3)
    back3_button.place(x=280,y=300)

    awindow.mainloop()

    
    
def remove(id):
        a="DELETE FROM empd WHERE ID=?"
        data=(id,)
        c=con.cursor()
        c.execute(a,data)
        con.commit()
        print("Employee deleted successfully\n")
        proceed()
        
def Remove():
    window2.destroy()
    global rwindow
    rwindow = tk.Tk()
    rwindow.geometry("400x350")
    rwindow.title("Employee Database-Admin Portal")

    head = tk.Label(rwindow, text="\nRemove Employee Details\n", font=("Arial", 20)).pack()
    
    def to_rem() :
        
        try:
            id=idtk.get()     
            id=int(id)
            
        except:
            print("ID must be integer only!!")
            tk.messagebox.showwarning("Warning", "ID must be in integer format!!")
            Remove()
            
        if check(id)==False:
            print("Employee does not exist!!")
            tk.messagebox.showwarning("Warning", "Employee does not exist!!")
        
    
        else:
            tk.messagebox.showinfo("Information", "Employee removed successfully!!")
            rwindow.destroy()
            remove(id)
            
                
    def back3() :
        rwindow.destroy()
        proceed()
        
    l_1 = tk.Label(rwindow, text=" Unique ID ").place(x=70, y=100)
    
    idtk = tk.Entry(rwindow, fg='blue', bg='white', width=30)
    idtk.place(x=140, y=100)

    Delete = tk.Button(rwindow, text="Delete", bg="green", fg="white", command=to_rem).place(x=190, y=300)
    
    back3_button = tk.Button(text="Back", bg="blue", fg="white", height=1, width=10, command=back3)
    back3_button.place(x=280,y=300)

    rwindow.mainloop()


    
    
def Display():
    
    window2.destroy()
    global dwindow
    dwindow = tk.Tk()
    dwindow.geometry("400x350")
    dwindow.title("Employee Database-Admin Portal")

    head = tk.Label(dwindow, text="\nEmployee Details\n", font=("Arial", 20)).pack()
                
    def to_proceed() :
        dwindow.destroy()
        proceed()
        
    def back3() :
        dwindow.destroy()
        proceed()
   
    def to_disp() :
        
        try:
            id=idtk.get()     
            id=int(id)
            
            if check(id)==False:
                print("Employee does not exist!!")
                tk.messagebox.showwarning("showwarning", "Employee does not exist!!")
        
    
            else:

                nametk,posttk,salarytk=display(id)
            
                l_2 = tk.Label(text="Name: " + nametk).place(x=70, y=150)
    
                l_3 = tk.Label(text="Post: " + posttk).place(x=70, y=200)
    
                l_4 = tk.Label(text="Salary: " + str(salarytk)).place(x=70, y=250)
                
            
    
        except:
            print("ID must be integer only!!")
            tk.messagebox.showwarning("Warning", "ID must be in integer format!!")
            
            
        
    
    def display(id):
        a="SELECT * FROM empd WHERE ID=?"
        data=(id,)
        c=con.cursor()
        x=c.execute(a,data)
       
        for i in x:
            print("Employ Id : ", i[0])
            print("Employ Name : ", i[1])
            print("Employ Post : ", i[2])
            print("Employ Salary : ", i[3])
            
            name=i[1]
            post=i[2]
            salary=i[3]
            return name,post,salary
    
    
    l_1 = tk.Label(dwindow, text=" Unique ID: ").place(x=70, y=100)
    
    idtk = tk.Entry(dwindow, fg='blue', bg='white', width=30)
    idtk.place(x=140, y=100)

    
    
    Ok = tk.Button(dwindow, text="OK", bg="green", fg="white", command=to_disp).place(x=190, y=300)
    
    back3_button = tk.Button(text="Back", bg="blue", fg="white", height=1, width=10, command=back3)
    back3_button.place(x=280,y=300)

    dwindow.mainloop()

    
    
    
def promote(id,npost,nsalary):
    
        a="SELECT POST,SALARY FROM empd WHERE ID=?"
        data=(id,)
        c=con.cursor()
        x=c.execute(a,data)
        
        post=npost
        salary=nsalary
        a="UPDATE empd set POST=?, SALARY=? WHERE ID=?"
        data=(post,salary,id)
        c.execute(a,data)
        con.commit()
        print("Salary changed successfully")
        pwindow.destroy()
        proceed()
        


def Promote():
    window2.destroy()
    global pwindow
    pwindow = tk.Tk()
    pwindow.geometry("600x600")
    pwindow.title("Employee Database-Admin Portal")

    head = tk.Label(pwindow, text="\nPromote Employee\n", font=("Arial", 20)).pack()
                
    def to_proceed() :
        pwindow.destroy()
        proceed()
        
    def back3() :
        pwindow.destroy()
        proceed()
   
    def to_dispns() :
        
        try:
            id=idtk.get()     
            id=int(id)
            
            if check(id)==False:
                print("Employee does not exist!!")
                tk.messagebox.showwarning("Warning", "Employee does not exist!!")
        
        except:
            print("ID must be integer only!!")
            tk.messagebox.showwarning("Warning", "ID must be in integer format!!")
            
        
        nametk,posttk,salarytk=display(id)
            
        l_2 = tk.Label(text="Name: " + nametk).place(x=70, y=150)
    
        l_3 = tk.Label(text="Post: " + posttk).place(x=70, y=200)
    
        l_4 = tk.Label(text="Salary: " + str(salarytk)).place(x=70, y=250)
   
    
        l_5 = tk.Label(pwindow, text="New Post ").place(x=70, y=300)
   
        posttk = tk.Entry(pwindow, fg='blue', bg='white', width=30)
        posttk.place(x=140, y=300)
    
        l_6 = tk.Label(pwindow, text="New Salary ").place(x=70, y=350)
    
        salarytk = tk.Entry(pwindow, fg='blue', bg='white', width=30)
        salarytk.place(x=140, y=350)
        
        def to_change():
            salary=salarytk.get()
            post=posttk.get()

            try:

                salary=int(salary)
                tk.messagebox.showinfo("Information", "Employee promoted successfully!!")    
                
                promote(id,post,salary)

            except:
                print("New Salary must be integer only!!")
                tk.messagebox.showwarning("Warning", "New Salary must be in integer format!!")
        

        Submit = tk.Button(pwindow, text="Promote", bg="green", fg="white", command=to_change).place(x=100, y=400)
        
                
            
        
    
    def display(id):
        a="SELECT * FROM empd WHERE ID=?"
        data=(id,)
        c=con.cursor()
        x=c.execute(a,data)
       
        for i in x:            
            name=i[1]
            post=i[2]
            salary=i[3]
            return name,post,salary
    
    
    l_1 = tk.Label(pwindow, text=" Unique ID: ").place(x=70, y=100)
    
    idtk = tk.Entry(pwindow, fg='blue', bg='white', width=30)
    idtk.place(x=140, y=100)

    
    Ok = tk.Button(pwindow, text="OK", bg="green", fg="white", command=to_dispns).place(x=350, y=100)
    
    back3_button = tk.Button(text="Back", bg="blue", fg="white", height=1, width=10, command=back3)
    back3_button.place(x=400,y=100)

    pwindow.mainloop()

        
        
    
main_screen()
```
